## 문제 1
아래 코드를 읽고 동작을 설명하시오.

````python
fishs = {
    '멸치': 10e-1,
    '광어': 3000,
    '고등어':500,
    '우럭': 1000,
    '참돔': 7000,
    '고래': 10e6,
}
````

````python
new_fishes = {}

for name, weight in fishs.items():
  if 1000 <= weight<= 8000:
    new_fishes[name] = weight

print(new_fishes)
````

````python
new1_fishes = { key:value for key, value in fishs.items() if 1000 <= value <=8000 }
print(new1_fishes)

new2_fishs = { name:weight for name, weight in fishs.items() if 1000 <= weight <= 8000 }
print(new2_fishs)
````

## 문제 4
딕셔너리형 데이터 `fishs`는 이름과 무게를 쌍으로 다섯 쌍의 요소로 되어 있습니다. 무게의 단위는 [g]입니다.

(1) 아래 코드의 동작을 설명하시오.
(2) `big_basket`에 들어 있는 우럭의 무게를 합산하면 몇 kg인를 구하는 코드를 추가로 작성하세요.

````python
import random

fishs = {
    '광어': 3000,     # 단위는 g
    '고등어':500,
    '우럭': 1000,
    '참돔': 7000,
    '고래': 10e6,
}

fish_types = ['광어', '고등어','참돔', '고래', '우럭']

big_basket = []

while( len(big_basket) <= 9 ):
  fish_type = fish_types[ random.randint(1, len(fish_types))-1 ]
  if fish_type != '고래':
    big_basket.append( (fish_type, fishs[fish_type]) )
````
