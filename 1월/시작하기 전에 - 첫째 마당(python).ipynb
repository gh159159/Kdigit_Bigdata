## 1-2. 최대값/최소값
- 리스트에서 최대,최소값 찾기


````python
#리스트에서 최대값 찾기
lst = [1, 2, 3, 9, 4, 5, 5]
print( max(lst) )

#리스트에서 최소값 찾기
print( min(lst) )
````

````python
#리스트에서 최대값 찾기
lst = [1, 2, 3, 9, 4, 5, 5]
print(lst.index( max(lst) ) )

#리스트에서 최소값 찾기
print( lst.index( min(lst) ) )
````

최대값이 2개 이상이라면

````python
#리스트에서 최대값 찾기
lst = [1, 2, 3, 9, 9, 4, 5, 5]
print(lst.index( max(lst) ) )

#리스트에서 최소값 찾기
print( lst.index( min(lst) ) )
````

## 2.참고 - 수치를 보다 전문적으로 다루기 위한 Numpy 모듈

````python
#numpy 배열에서 최대/최소값 찾기
import numpy as np
lst = [1, 2, 3, 9, 4, 5, 5]
narray = np.array(lst)
print( narray.max() )
print( narray.min() )
````

````python
#numpy 배열에서 최대/최소값의 인덱스스 찾기
import numpy as np
lst = [1, 2, 3, 9, 4, 5, 5]
narray = np.array(lst)
print( narray.max() )
print( narray.min() )
````

````python
#numpy 배열에서 최대/최소값의 인덱스스 찾기
import numpy as np
lst = [1, 2, 3, 9, 4, 5, 5]
narray = np.array(lst)
print( narray.argmax() )
print( narray.argmin() )
````

최소/최대값을 찾으려고 하는데, 최소 또는 최대값이 2개 이상이라면

````python
#numpy 배열에서 최대/최소값 찾기
import numpy as np
lst = [1, 2, 3, 9, 9, 4, 5, 5]
narray = np.array(lst)

print( np.where(narray == narray.max()) )
print( np.where(narray == narray.min()) )
````
