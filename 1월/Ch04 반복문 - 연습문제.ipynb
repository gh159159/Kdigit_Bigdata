```python
lst = ['홍길동', '전우치', 25, 32]
rlst = ['홍길동님', '전우치님', 26, 33]
```

####풀이 9-1

````python
lst = ['홍길동', '전우치', 25, 32]
print(lst)

for idx in range(len(lst)):
  if type(lst[idx]) == str:
    lst[idx] += '님'
  else:
    lst[idx] += 1

print(lst)
````

####풀이 9-2

````python
lst = ['홍길동', '전우치', 25, 32]
print(lst)

rlst = []

for element in lst:
  rlst.append( element+'님' if type(element) == str else element+1 )

print(rlst)
````

####풀이 9-3

##문제 18
아래 코드에 보이는 리스트, `prices`의 각 요소는 그액을 달러 단위 문자열로 표시되어 있습니다. 리스트 `prices`의 모든 요소를 다 더하면 몇 달러인지와 원화로 환산하면 얼마인지 출력하는 코드를 작성하세요. (2023년 5월 28일 기준 1달러는 1,328원입니다.)

```python
prices = ['2,000$', '3,000$', '5,000$']
exchange_rate = 1328
```

###풀이 18-1

````python
prices = ['2,000$', '3,000$', '5,000$']
exchange_rate = 1328

d_total = 0

for price in prices:
  dollar = int(price.replace('$', '').replace(',', ''))
  d_total +=  dollar

total = d_total * exchange_rate


print(d_total,'$')
print(total,'\\')
````

###풀이 18-2

````python
prices = ['2,000$', '3,000$', '5,000$']
exchange_rate = 1328

wprice = []
d_total = 0

for price in prices:
  dollar = int(price.replace('$', '').replace(',', ''))
  d_total +=  dollar
  wprice.append( dollar )

total = d_total * exchange_rate


print(str(d_total)+'$')
print(str(total)+'\\')
````

###풀이 18-3

````python
prices = ['2,000$', '3,000$', '5,000$']
exchange_rate = 1328

wprice = []
d_total = 0

for price in prices:
  dollar = int(price.replace('$', '').replace(',', ''))
  d_total +=  dollar
  wprice.append( dollar )

total = d_total * exchange_rate


print(format(d_total,',')+'$')
print(format(total, ',')+'\\')
````

#### 18-4

## 문제 28
길이가 23cm이상이고 무게는 250g이상인 생선은 '쿨러에 담는다'고 출력하고 기준에 미달되는 생선은 '방생'이라고 출력합니다.

```python
fish_length = [14, 15.5, 13.5, 23.5, 28.5, 26.0, 12, 32, 29, 14.5]
fish_weight = [100, 150, 140, 250, 300, 280, 80, 315, 220, 90]
```

* (1) 반복 동작을 `for`문으로 구현하십시오.
* (2) 반복 동작을 `while`문으로 구현하십시오.

###풀이 28-(1)-1

````python
fish_length = [14, 15.5, 13.5, 23.5, 28.5, 26.0, 12, 32, 29, 14.5]
fish_weight = [100, 150, 140, 250, 300, 280, 80, 315, 220, 90]

# for문 활용
for idx in range(len(fish_length)):
  length = fish_length[idx]
  weight = fish_weight[idx]

  if length >= 23 and weight >= 250:
    print('쿨러에 담는다. ({}, {})'.format(length, weight))
  else:
    print('방생')
````

###풀이 28-(1)-2
`zip()`을 이용하여 `fish_length`와 `fish_weight`를 하나로 묶어서 처리

````python
fish_length = [14, 15.5, 13.5, 23.5, 28.5, 26.0, 12, 32, 29, 14.5]
fish_weight = [100, 150, 140, 250, 300, 280, 80, 315, 220, 90]

# for문 활용
for length, weight in zip(fish_length, fish_weight):
  if length >= 23 and weight >= 250:
    print('쿨러에 담는다. ({}, {})'.format(length, weight))
  else:
    print('방생')
````

###풀이 28-(2)-1

##문제 35
변수 $x$의 범위는 1부터 15까지 양의 정수입니다.
$$y = 2x + 3$$
변수 $y$의 값을 리스트 형태로 보관하고 싶습니다. 그런데 이 때 $x$의 값 5일 때와 11일 때는 제외하고 싶습니다.




###풀이 35-1


````python
y = []
for x in range(1, 16):
  if x == 5 or x == 11:
    continue
  y.append(2*x+3)

print(y)
````

````python
y = []
for x in range(1, 16):
  if x != 5 and x != 11:
    y.append(2*x+3)

print(y)
````

`if x != 5 and x != 11:`에서 `x`가 1일 때 ,`True and True` 따라서 조건은 `True`가 됩니다.

|`x`|`x!=5`|`x!=11`|`x!=5 and x!=11`|
|:-:|:-:|:-:|:-:|
|1|`True`|`True`|`True`|
|2|`True`|`True`|`True`|
|3|`True`|`True`|`True`|
|...|...|...|...|
|5|`Fasle`|`True`|`False`|
|...|...|...|...|
|11|`True`|`False`|`False`|
|...|...|...|...|
|15|`True`|`True`|`True`|




###풀이 35-2


````python
y = [ 2*x+3 for x in range(1, 16) if x != 5 and x != 11]
print(y)
# 2*5 + 3 = 13
# 2*11 + 3 = 25
````

### 풀이 35-3

````python
y = [2*x + 3 for x in range(1, 16) if x not in [5, 11]]
print(y)
````

```python
numbers = [1, 2, 3, 4, 5, 6]

print(next(r_num))
print(next(r_num))
print(next(r_num))
print(next(r_num))
print(next(r_num))
print(next(r_num))
```
