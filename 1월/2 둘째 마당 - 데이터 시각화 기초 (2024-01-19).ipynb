#UNIT 05 내 생일의 기온 변화를 그래프로 그리기

###1. 데이터에 질문하기
매년 여러 분의 생일 날 서울의 최고 기온은 어땠을가요?

#### 데이터 읽어 오기

````python
import csv
f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)           # 헤더 제거

for row in data:
  print(row[-1])      # 최고 기온값만 출력

f.close()
````

####데이터를 리스트에 저장하기
누락 데이터를 제외 하고 최고 기온 값만 추출하여 리스트 형 데이터를 만들겠습니다. 아래 코드에서 ```max_temp_list```는 리스트 형 변수로 매일 매일의 최고 기온 값을 보관하기 위한 변수입니다.


````python
import csv
f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)            # 헤더 제거

max_temp_list =[]     # 최고 기온 데이터를 저장하기 위해 미리 빈 리스트 생성

for row in data:
  # 누락된 데이터가 아닐 경우(데이터가 있을 경우)
  if row[-1] != '':
    # row[-1]은 문자열 데이터임으로 float() 함수를 사용하여 숫자형 데이터로 변환 후 저장
    max_temp_list.append(float(row[-1]))

f.close()

print(max_temp_list)
print( len(max_temp_list) )
````

####데이터 시각화하기
매일 매일의 최고 기온을 ```matplotlib``` 모듈의 ```plot()```함수로 그려보겠습니다.

````python
import matplotlib.pyplot as plt
plt.plot(max_temp_list, 'r')
plt.show()
````

눈으로 봐서 값의 변화 추이를 알아보기 어려우니 그림(figure)의 크기를 키워 보겠습니다.

````python
plt.figure(figsize=(16, 4))
plt.plot(max_temp_list, 'r')
plt.show()
````

####날자 데이터 추출하기
'부산.csv' 파일에서 구한 데이터는 아래와 같은 형태의 데이터입니다.
```python
['\t2009-06-14', '159', '20.2', '18.2', '23.8']
```
위 리스트에서 날자에 해당하는 정보는 첫 번째 요소로 단순 문자열로 기록되어 있습니다.

```python
'\t2009-06-14'
```
에서 년, 월, 일 정보를 추출하려고 합니다. 먼저 ```replace()```함수를 사용하여 ```'\t2009-06-14'```에서 ```\t```을 제거하겠습니다.


````python
date = '\t2009-06-14'
date = date.replace('\t', '')
print(date)
````

그리고 ```'2009-06-14'```의 문자열은 ```split()```함수를 이용하여 년, 월, 일을 분리하겠습니다.

````python
ymd = date.split('-')
print(ymd)
````

````python
print(ymd[0])   # 년 : year
print(ymd[1])   # 월 : month
print(ymd[2])   # 일 : day
````

1년 중 가장 더운 달은 8월 달의 최고 기온 데이터만 추출하여 그래프로 그려보겠습니다.

````python
import csv
f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)

max_temp_list =[]

for row in data:
  if row[-1] != '':                         # 누락된 데이터가 아닐 경우(데이터가 있을 경우)
    if row[0].split('-')[1] == '08':          # ['\t2009-06-14', '159', '20.2', '18.2', '23.8']
      max_temp_list.append(float(row[-1]))    # row[-1]은 문자열 데이터임으로 float() 함수를 사용하여 숫자형 데이터로 변환 후 저장

f.close()

plt.figure(figsize=(16, 4))
plt.plot(max_temp_list, 'r')
plt.show()
````

위 그래프를 보면서 어떤 의미를 찾기가 쉽지 않네요. 그렇다면 매년 자기 자신의 생일 날 최고 기온의 변화를 살펴보겠습니다.

````python
import csv
f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)            #해더 제거
# 생일 날의 최고 기온 데이터를 년도 순으로 max_temp_lst에 모음
max_temp_list =[]
for row in data:
  if row[-1] != '':                 # 최고 기온 데이터가 ''이 아닐 때
    ymd = row[0].split('-')
    if ymd[1] == '01' and ymd[2] =='24':
      max_temp_list.append(float(row[-1]))  # '최고 기온'은 문자열 데이터임으로 float() 함수를 사용하여 숫자형 데이터로 변환 후 저장

f.close()
# 그림 그리기
plt.figure(figsize=(16, 4))
plt.plot(max_temp_list, 'b.-')
plt.title('max. temp of January 24 every year')
plt.ylabel('temp')
plt.grid()
plt.show()
````

#####Quiz
매년 생일 날의 최고 기온과 최저 기온을 하나의 그림(figure)에 나타내는 코드를 작성하세요.

````python
import csv
f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)
max_temp_list =[]
min_temp_list = []

for row in data:
  if row[-1] != '' and row[-2] != '':                                                             # 누락된 데이터가 아닐 경우(데이터가 있을 경우)
    if row[0].split('-')[1] == '01' and row[0].split('-')[2] =='24':
      max_temp_list.append(float(row[-1]))                                      # row[-1]은 문자열 데이터임으로 float() 함수를 사용하여 숫자형 데이터로 변환 후 저장
      min_temp_list.append(float(row[-2]))

f.close()

plt.figure(figsize=(16, 4))
plt.plot(max_temp_list, 'r', label='max. temp')
plt.plot(min_temp_list, 'b', label='min. temp')
plt.title('max, min temp. of January 24 every year')
plt.ylabel('temp')
plt.legend()
plt.grid()
plt.show()
````

````python
import csv
f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)           # 헤더 제거
data_lst = list(data)
f.close()
````

````python
# 결측치 제거
data_lst = [data for data in data_lst if data[-1] != '']


data_dic ={}
# 년도별 데이터로 분리
for day_data in data_lst:
  ymd = day_data[0].replace('\t', '')       # 예, ymd = '1907-10-01'
  year = int(ymd.split('-')[0])             # 예, 1907
  max_temp = float(day_data[-1])            # 예,

  if year in data_dic:
    data_dic[year].append((ymd, max_temp))
  else:
    data_dic[year] = [(ymd, max_temp)]        #예 [('1907-10-01', 2.3)]
````

````python
#data_dic.keys()
#data_dic[1907]
````

````python
result = [[],[]]
for year in data_dic.keys():
  if year == 1907 or year == 2024:
    continue
  data_a_year = data_dic[year]


  temp = [ day_data[1] for day_data in data_a_year if day_data[0].split('-')[1] == '08' ]
  if not temp:
    continue
  else:
    mean_temp = sum(temp)/len(temp)
    result[0].append(year)
    result[1].append(mean_temp)
````

````python
plt.plot(result[0], result[1])
plt.xlabel('year')
plt.title('mean(max temp @ Aug.)')
plt.grid()
plt.show()
````

````python
import csv
f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)           # 헤더 제거
data_lst = list(data)
f.close()
````

````python
#######################################################################
# 데이터 (사)전처리

# 결측치 제거
# '최고 기온' 값이 없는 날의 데이터는 제거
data_lst = [data for data in data_lst if data[-1] != '']

# 데이터 정제
# 필요 데이터(날자, 최고 기온)만 추출하고
# 날자 데이터에서 '\t' 제거 그리고 문자열로 표현된 '최고 기온'숫자로 변경
# 예 ['1907-10-01', 23.1]
data_lst = [ (data[0].replace('\t', ''), float(data[-1])) for data in data_lst if data[-1] != '']
# data_lst -> [  ('1907-10-01', 23.1),  ('1907-10-02', 22.1), .... ]

# 데이터를 정리
each_year_data_dic = {}

for ymd, max_temp in data_lst:
  year = int(ymd.split('-')[0])

  if year in each_year_data_dic:
    each_year_data_dic[year].append((ymd, max_temp))
  else:
    each_year_data_dic[year] = [(ymd, max_temp)]        #예 [('1907-10-01', 2.3)]
````

````python
# for 문에서 반복이 한 번 실행될 때 마다
# 1년 치 데이터(each_year_data_dic[1907])를 뽑아서
result = [[],[]]
for year in each_year_data_dic.keys():
  data_a_year =  each_year_data_dic[year]
  max_temp_lst = [ max_temp for ymd, max_temp in data_a_year if ymd.split('-')[1] == '08' ]
  if not max_temp_lst:
    continue
  else:
    mean_temp = sum(max_temp_lst)/len(max_temp_lst)
    result[0].append(year)
    result[1].append(mean_temp)
````

````python
plt.plot(result[0], result[1])
plt.xlabel('year')
plt.title('mean(max temp @ Aug.)')
plt.grid()
plt.show()
````

````python
import csv
f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)           # 헤더 제거
data_lst = list(data)
f.close()
````

````python
#######################################################################
# 데이터 (사)전처리

# 결측치 제거
# '(일) 평균 기온' 값이 없는 날의 데이터는 제거
data_lst = [data for data in data_lst if data[-2] != '']
````

````python
data_lst[0]
````

````python

# 데이터 정제
# 필요 데이터(년도, 일 평균 기온)만 추출하고
# 날자(년월일) 데이터에서 '\t' 제거하고 년도를 요소로 그리고 문자열로 표현된 '일 평균 기온'을 숫자로 변경하여 요소로,
# 두 가지를 요소로 하는 튜플
# 예 [('1907', 23.1)]
data_lst = [ ( int(a_day_data[0].replace('\t', '').split('-')[0]), float(a_day_data[-2])) for a_day_data in data_lst]
# data_lst -> [  (1907, 23.1),  (1907, 22.1), .... ]
````

````python

# 데이터를 정리
# key는 년도(정수)
# value는 [일 평균 기온, 일 평균 기온, 일 평균 기온, ....]
# 예) 1907:[23.1, 23.3, ....]
each_year_data_dic = {}

for year, mean_temp in data_lst:
  if year in each_year_data_dic:
    each_year_data_dic[year].append(mean_temp)
  else:
    each_year_data_dic[year] = [mean_temp]        #예 1907:[2.3, 2.4, ...]

mean_temp_year =[]
years =[]
###########################################
for year in each_year_data_dic.keys():
  day_count = len(each_year_data_dic[year])
  if day_count < 360:
    continue
  mean_temp_year.append( sum(each_year_data_dic[year])/day_count )
  years.append(year)
````

````python
plt.plot(years,mean_temp_year)
plt.xlabel('year')
plt.title('Mean Temp.')
plt.grid()
plt.show()
````
