```python
import csv
import matplotlib.pyplot as plt

f = open('gender_202306.csv', encoding='cp949')
data = list( csv.reader(f) )[1:]
f.close()
```

```python
# data : 파일에서 읽어온 데이터
# pop_info : 한 지역에 대한 연령대별 남여 인구수, 리스트형
# pop_info[0]:행정구역(이름)
busan_pop = [ pop_info for pop_info in data if '2600000000' in pop_info[0] ][0]
seoul_pop = [ pop_info for pop_info in data if '1100000000' in pop_info[0] ][0]
```

```python
%page busan_pop
```

```python
# busan_pop[3:14] : 연령대 별 남자 인구수
# busan_pop[16:]  : 연령대 별 여자 인구수
busan_pop = busan_pop[3:14] + busan_pop[16:]
# 문자열로 표현된 수에서 쉼표 제거 후 정수로 변환
busan_pop = [ int(count.replace(',','')) for count in busan_pop ]

seoul_pop = seoul_pop[3:14] + seoul_pop[16:]
seoul_pop = [ int(count.replace(',','')) for count in seoul_pop ]
```

```python
%page busan_pop
```

```python
# 튜플 언팩
busan_m_pop, busan_w_pop = busan_pop[:11], busan_pop[11:]
seoul_m_pop, seoul_w_pop = seoul_pop[:11], seoul_pop[11:]
```

```python
%page busan_m_pop
```

```python
# 남자 인구수 리스의 각 요소 값의 부호만 음수로
busan_m_pop = [ -1*count for count in busan_m_pop ]
seoul_m_pop = [ -1*count for count in seoul_m_pop ]
```

```python
plt.figure(figsize=(16,9))

plt.subplot(2,1,1)
plt.title('Seoul')
plt.barh(range(11), seoul_m_pop, label='남자 인구수')
plt.barh(range(11), seoul_w_pop, label='여자 인구수')
plt.xlim([-800000, 800000])
plt.grid()

plt.subplot(2,1,2)
plt.title('Busan')
plt.barh(range(11), busan_m_pop, label='남자 인구수')
plt.barh(range(11), busan_w_pop, label='여자 인구수')
plt.xlim([-800000, 800000])
plt.grid()

plt.show()
```

여러 코드 셀로 나누어진 코드를 하나의 코드셀로 모으겠습니다.

```python
import csv
import matplotlib.pyplot as plt

f = open('gender_202306.csv', encoding='cp949')
data = list( csv.reader(f))[1:]
f.close()
#------------------------------------------------------------------------------
busan_pop = [ pop_info for pop_info in data if '2600000000' in pop_info[0] ][0]
seoul_pop = [ pop_info for pop_info in data if '1100000000' in pop_info[0] ][0]

#------------------------------------------------------------------------------
busan_pop = busan_pop[3:14] + busan_pop[16:]
busan_pop = [ int(count.replace(',','')) for count in busan_pop ]
seoul_pop = seoul_pop[3:14] + seoul_pop[16:]
seoul_pop = [ int(count.replace(',','')) for count in seoul_pop ]
#------------------------------------------------------------------------------
busan_m_pop, busan_w_pop = busan_pop[:11], busan_pop[11:]
seoul_m_pop, seoul_w_pop = seoul_pop[:11], seoul_pop[11:]

#------------------------------------------------------------------------------
busan_m_pop = [ -1*count for count in busan_m_pop ]
seoul_m_pop = [ -1*count for count in seoul_m_pop ]

#------------------------------------------------------------------------------
plt.figure(figsize=(16,9))

plt.subplot(2,1,1)
plt.title('Seoul')
plt.barh(range(11), seoul_m_pop, label='남자 인구수')
plt.barh(range(11), seoul_w_pop, label='여자 인구수')
plt.xlim([-800000, 800000])
plt.grid()

plt.subplot(2,1,2)
plt.title('Busan')
plt.barh(range(11), busan_m_pop, label='남자 인구수')
plt.barh(range(11), busan_w_pop, label='여자 인구수')
plt.xlim([-800000, 800000])
plt.grid()

plt.show()
```
