```python

```


````python
class Tiguan():
  def __init__(self, name):
    self.name = name

tiguan = Tiguan('쫑')

print(tiguan.name)
````

```python

```
위 statement(문장)이 실행되면

###객체(Object)
학생 성적 관리 프로그램을 만든다면 무엇이 필요할까요? 학생 이름, 학번, 과목별 성적 등이 필요합니다. 다음과 같이 이런 사항들을 딕셔너리로 표현해 보겠습니다.





````python
# file name : object_1_basic.py

students =[
    {'name':'김성필', 'korean':93, 'math':83, 'english':92, 'science':93},
    {'name':'구숙정', 'korean':90, 'math':80, 'english':95, 'science':87},
    {'name':'장만옥', 'korean':92, 'math':79, 'english':93, 'science':88},
    {'name':'박혜진', 'korean':88, 'math':83, 'english':81, 'science':81},
    {'name':'조혜정', 'korean':84, 'math':89, 'english':86, 'science':85},
    {'name':'김재벌', 'korean':95, 'math':90, 'english':94, 'science':90}
]

print('이름', '총점', '평균', sep='\t')

for student in students:
  # 총점 계산
  score_sum = student['korean']         \
              + student['math']         \
              + student['english']      \
              + student['science']
  # 평균 계산
  score_average = score_sum / 4
  print(student['name'], score_sum, score_average, sep='\t')
  #print('{}\t{:4d}\t{:3.2f}'.format(student['name'], score_sum, score_average))
````

딕셔너리로 학생 관련 정보를 표현했고 이를 각 요소로 해서 전체 데이터는 리스트로 만들었습니다. 그런데 딕셔너리로 한 학생에 대한 정보를 하나씩 만드니 조금 복잡하고 귀찮네요. 딕셔너리 형태인 학생에 대한 정보를 타이핑하다가 키(key)를 잘못(오타)입력하기도 했습니다.

<br>

딕셔너리를 생성하는 코드를 다음과 같은 형태의 함수로 만들면 어떨까요? 딕녀너리 형태로 타이핑해서 학생 정보를 입력할 때 보다 쉽고 키를 잘못 입력하는 실수를 하지 않게 됩니다.

````python
# 각 학생에 대한 정보를 딕셔너리형태로 반환하는 함수를 선언합니다.
def create_student(name, korean, math, english, science):
  return {
          'name':name,
          'korean':korean,
          'math': math,
          'english':english,
          'science':science
          }

# 각 학생에 대한 정보를 요소로 하는 리스트를 생성합니다.
students = [
            create_student('김성필', 93, 83, 92, 93),
            create_student('구숙정', 90, 80, 95, 87),
            create_student('장만옥', 92, 79, 93, 88),
            create_student('박혜진', 88, 83, 81, 81),
            create_student('조혜정', 84, 89, 86, 85),
            create_student('이재벌', 95, 90, 94, 90),
            ]

print('이름', '총점', '평균', sep='\t')

for student in students:
  # 한 학생의 총점을 계산
  score_sum = student['korean']         \
              + student['math']         \
              + student['english']      \
              + student['science']

  # 한 학생의 평균을 구함.
  score_average = score_sum / 4

  # 관심 정보 출력
  print(student['name'], score_sum, score_average, sep='\t')
````

실행 결과는 이전과 같습니다.

<br>

현재 <font color=blue> **'총점'과 '평균'을 구하는 처리는 학생을 대상으로만 이루어집니다.** </font>따라서 학생 이름을 매개 변수 받는 함수를 만들어서 해당 처리를 하면 코드가 조금 멋져 보일 것 같습니다.

````python
####################################################################
### - 학생 데이터를 처리하는 함수들

# 각 학생에 대한 정보를 딕셔너리형태로 반환하는 함수를 선언합니다.
def create_student(name, korean, math, english, science):
  return {                      # 딕셔너리형 데이터를 반환
          'name':name,
          'korean':korean,
          'math': math,
          'english':english,
          'science':science
          }

# 한 학생 점수를 처리하는 함수들 -------------------------------

# 1. 총점 계산 함수
def get_sum_student(student):
  # 한 학생의 총점을 계산
  # 'student'는 딕셔너리 형 데이터군요.
  return student['korean'] + student['math'] \
        + student['english'] + student['science']

#2. 평균 점수 계산하는 함수
def get_average_student(student):
  return get_sum_student(student) / 4

#3. 학생 (성적)정보 추출하는 함수
def get_info_student(student):
  return '{}\t{:3d}\t{:3.2f}'.format(
                              student['name'],
                              get_sum_student(student),
                              get_average_student(student)
                            )
 ################################################################

 ################################################################
 ### - 학생 데이터
 # 각 학생에 대한 정보를 요소로 하는 리스트를 생성합니다.
students = [
            create_student('김성필', 93, 83, 92, 93),
            create_student('구숙정', 90, 80, 95, 87),
            create_student('장만옥', 92, 79, 93, 88),
            create_student('박혜진', 88, 83, 81, 81),
            create_student('조혜정', 84, 89, 86, 85),
            create_student('이재벌', 95, 90, 94, 90),
            ]

#################################################################
### 학생 데이터와 학생 데이터를 처리하는 함수를 사용하는 부분 ###

print('이름', '총점', '평균', sep='\t')
for student in students:
  # 한 학생에 대한 정보 출력
  print( get_info_student(student) )
````

위와 같이 코드를 작성함으로써 기능이나 역활을 기준으로 코드를 구분할 수 있게 되었습니다. 학생을 대상으로하는 기능은 윗 부분에 코딩하였고 학생에 대한 데이터를 사용하는 부분은 아래에 코딩했습니다. 이렇게 만들면 '학생에 관련된 기능(함수)'를 모아서 별도의 모듈로 분리하여 관리할 수도 있겠습니다. 위 코드는 학생을 '대상'으로한 프로그램 코드에서 기능(함수), 데이터를 분리하여 코딩한 것이라고 말할 수 있겠습니다.

<br>

위와 같은 구조로 코딩을 하다 보니 개발자들은 클래스(class)라는 개념내지는 구조를 만들게 되었습니다.

`isinstance()` 함수는 다양하게 활용할 수 있습니다. 간단한 예로 하나의 리스트 내부에 여러 종류의 인스턴스가 들어 있을 때, 인스턴스를 구분하여 해당 인스턴스의 속성과 기능을 사용할 때 이용할 수 있습니다.

다음 코드를 살펴보겠습니다. `Student`와 `Teacher`라는 클래스를 생성하고 `classroom` 리스트에 인스턴스를 넣었습니다. 그리고 이 리스트를 반복문으로 처리하려고 할 때 각 요소가 `Student`클래스의 인스턴스인지 `Teacher` 클래스의 인스턴스 확인하여 적절한 멤버 함수를 호출합니다.


````python
# 학생 클래스 선언
class Student:
  def study(self):
    print('인생 공부를 합니다.')

class Teacher:
  def teach(self):
    print('지식을 가르칩니다. 때때로 지혜도.')

classroom = [Student(), Student(), Teacher(), Student(), Student()]

for person in classroom:
  if isinstance(person, Student):
    person.study()
  elif isinstance(person, Teacher):
    person.teach()
````

일반적으로 객체 지향 프로그래밍은 모든 데이터를 클래스로 구현합니다. 이러한 데이터를 관리할 때 종류별로 리스트를 따로 만들고 활용해야 한다고 생각하는 경우가 많은데요, `isinstance()` 함수를 사용하면 이처럼 하나의 리스트로도 여러 종류의 데이터를 다룰 수 있습니다.

###특수한 이름의 메소드(멤버함수)
우리가 만든 Student 클래스를 기반으로 객체를 만들고 객체 뒤애 .(마침표)를 입력해서 비주얼 스튜디오 코드의 자동 완성 기능을 살펴보면 우리가 만들지 않은 함수들이 있다는 것을 확인할 수 있습니다.

````python
class Student:
    def __init__(self, name, korean, math, english, science):
        self.name = name
        self.korean = korean
        self.math = math
        self.english = english
        self.science = science
    def get_sum(self):
      return self.korean + self.math + self.english + self.science

    def get_average(self):
        return self.get_sum()/4

    # __str__() 함수 선언
    def __str__(self):
        return '{}\t{}\t{}'.format(self.name, self.get_sum(), self.get_average())

students = [
    Student('김성필', 93, 83, 92, 93),
    Student('구숙정', 90, 80, 95, 87),
    Student('장만옥', 92, 79, 93, 88),
    Student('박혜진', 88, 83, 81, 81),
    Student('조혜정', 84, 89, 86, 85),
    Student('이재벌', 95, 90, 94, 90),
]

print('이름', '총점', '평균', sep='\t')
for student in students:
    # str()함수의 매개변수로 student 인스턴스를 전달하면
    # student 인스턴스의 __str__() 함수가 호출됨
    print( str(student) )
````

`__str__()`과 같이 사전에 (멤서함수로 쓰겠다고 예약된) 이름을 사용하고 `__<이름>__`형태로 클래스에서 메소드(멤버 함수)를 정의하면 `인스턴스.메소드()` 형태가 아닌 `메소드(인스턴스)` 형태로 클래스 내부에 정의된 `__<이름>__()` 함수를 호출 할 수 있습니다.

<br>

이렇게 사용할 수 있도록 예약된 이름을 살펴 보면 다음과 같습니다.

<br>
<center>

|이름|의미|한글 설명|
|:-:|:-:|:-:|
|eq|equal|같다|
|ne|not equal|같지 않다|
|gt|greater than ~|~ 보다 크다|
|ge|greater than or equal|크거나 같다|
|lt|less than ~|~ 보다 작다|
|le|less than or equal|작거나 같다|

</center>


이러한 함수를 활용하여 학생들의 성적을 비교할 수 있도록 코드를 수정해 보겠습니다. 위에서 봤던 `__str__()`과 양상이 다릅니다.

````python
class Student:
  def __init__(self, name, korean, math, english, science):
    self.name = name
    self.korean = korean
    self.math = math
    self.english = english
    self.science = science

  def get_sum(self):
    return self.korean + self.math + self.english + self.science

  def get_average(self):
    return self.get_sum()/4

  # __str__() 함수 선언
  def __str__(self):
    return '{}\t{}\t{}'.format(self.name, self.get_sum(), self.get_average())

  # 연산자 재정의
  def __eq__(self, student):
    return self.get_sum() == student.get_sum()
  def __ne__(self, student):
    return self.get_sum() != student.get_sum()
  def __gt__(self, student):
    return self.get_sum() > student.get_sum()
  def __ge__(self, student):
    return self.get_sum() >= student.get_sum()
  def __lt__(self, student):
    return self.get_sum() < student.get_sum()
  def __le__(self, student):
    return self.get_sum() <= student.get_sum()


students = [
    Student('김성필', 93, 83, 92, 93),
    Student('구숙정', 90, 80, 95, 87),
    Student('장만옥', 92, 79, 93, 88),
    Student('박혜진', 88, 83, 81, 81),
    Student('조혜정', 84, 89, 86, 85),
    Student('이재벌', 95, 90, 94, 90)
]


print('이름', '총점', '평균', sep='\t')
for student in students:
  # str()함수의 매개변수로 student 인스턴스를 전달하면
  # student 인스턴스의 __str__() 함수가 호출됨
  print( str(student) )

print('--- 구분선 ---')

print('{}\t: student[0] == student[1]'.format( students[0] == students[1] ) )
print('{}\t: student[0] != student[1]'.format( students[0] != students[1] ) )

print('{}\t: student[0] > student[1]'.format( students[0] > students[1] ) )
print('{}\t: student[0] >= student[1]'.format( students[0] >= students[1] ) )

print('{}\t: student[0] < student[1]'.format( students[0] < students[1] ) )
print('{}\t: student[0] <= student[1]'.format( students[0] <= students[1] ) )

print('--- 구분선 ---')
s = Student('김성필', 93, 83, 92, 93)
print( s.__str__() )

````

######클래스 함수 만들기

```python
class 클래스_이름:

```

클래스 함수의 첫 번째 매개변수에는 클래스 자체가 들어옵니다. 일반적으로 `cls`(그냥 클래스라고 읽으면 됩니다)라는 이름의 변수로 선언하며 이렇게 만들어진 클래스 함수는 다음과 같이 사용하니다.

######클래스 함수 호출하기

```python

```

그럼 이제 간단하게 클래스 함수를 만들고 클래스 함수를 활용해 보도록 하겠습니다. `students`라는 학생 리스트를 아예 클래스 내부에 만들어 버리고 학생 리스트를 전부 출력하는 `Student.print()` 함수를 만들겠습니다.


````python
class Student:
  # 클래스 변수
  count = 0
  students =[]

  # 클래스 함수
  @classmethod
  def print(cls):
    print('----- 학생 목록 -----')
    print('이름\t총점\t평균')
    for student in cls.students:
      print(str(student))


  #인스턴스 함수
  def __init__(self, name, korean, math, english, science):
    self.name = name
    self.korean = korean
    self.math = math
    self.english = english
    self.science = science

    Student.count += 1          # 클래스 변수에 접근
    Student.students.append(self)

  def get_sum(self):
    return self.korean + self.math + self.english + self.science

  def get_average(self):
    return self.get_sum()/4

  # __str__() 함수 선언
  def __str__(self):
    return '{}\t{}\t{}'.format(self.name, self.get_sum(), self.get_average())


Student('김성필', 93, 83, 92, 93),
Student('구숙정', 90, 80, 95, 87),
Student('장만옥', 92, 79, 93, 88),
Student('박혜진', 88, 83, 81, 81),
Student('조혜정', 84, 89, 86, 85),
Student('이재벌', 95, 90, 94, 90)

Student.print()
````

클래스 내부의 `__radius`를 클래스 밖에서 접근하게 되면 어떻게 될까요? 클래스 밖에서 `__radius`에 값을 읽거나 값을 쓰려고 하면 어떻게 될까요?

````python
# 모듈 가져오기
import math

# 클래스 선언
class Circle:
  def __init__(self, radius):
    self.__radius = radius

  def get_circumference(self):
    return 2 * math.pi * self.__radius

  def get_area(self):
    return math.pi * self.__radius**2

circle = Circle(10)
print('원의 반지름:', circle.__radius)
````

"AttributeErrror: 'Circle' object has no attribute '\_\_radius'"라는 에러 메시지가 출력되었습니다. 클래스 외부에서 `__radius`를 사용하려고 할 때 그런 속성이 없다는 오류 메시지 입니다. 이처럼 속성을 선언할 때 속성 이름 앞에 `__`를 붙이기만 하면 외부에서 사용할 수 없는 변수가 됩니다.

**<ins>게터와 세터</ins>**
그런데 프로그램 실행 도중에 원의 반지름 값을 바꾸고 싶다면 어떻게 해야할까요? 클래스 외부에서 직접 `__radius``속성에 접근할 수 없기 때문에 간접적으로 접근할 수 있는 방법을 찾아야 합니다.

<br>

이때 사용되는 것이 <font color=red>게터</font>(getter)와 <font color=red>세터</font>(setter)입니다. 게터는 프라이빗 속성의 값을 추출하기 위해 세터는 브라이빗 속성의 값을 변경하기 위해 간접적으로 접근할 수 있도록 해주는 함수입니다.

````python
import math

class Circle:
  def __init__(self, radius):
    self.__radius = radius

  def get_circumference(self):
    return 2 * math.pi * self.__radius

  def get_area(self):
    return math.pi * self.__radius**2

  # 게터와 세터 선언
  def get_radius(self):
    return self.__radius
  def set_radius(self, value):
    self.__radius = value


circle = Circle(10)
print('원의 반지름:', circle.get_radius())          # (간접적으로) __radius 속성에 접근
print('원의 둘례:', circle.get_circumference())
print('원의 넓이:', circle.get_area())

print()

circle.set_radius(2)
print('원의 반지름:', circle.get_radius())          # (간접적으로) __radius 속성에 접근
print('원의 둘례:', circle.get_circumference())
print('원의 넓이:', circle.get_area())
````

````python
import math

class Circle:
  def __init__(self, radius):
    self.__radius = radius

  def get_circumference(self):
    return 2 * math.pi * self.__radius

  def get_area(self):
    return math.pi * self.__radius**2

  # 게터와 세터 선언
  def get_radius(self):
    return self.__radius
  def set_radius(self, value):
    if value < 0:
      raise ValueError("반지름 값은 양수이어야 합니다.")
    self.__radius = value



circle = Circle(10)
print('원의 반지름:', circle.get_radius())          # (간접적으로) __radius 속성에 접근
print('원의 둘례:', circle.get_circumference())
print('원의 넓이:', circle.get_area())

print()

circle.set_radius(-2)
````
````python
print('원의 반지름:', circle.get_radius())          # (간접적으로) __radius 속성에 접근
print('원의 둘례:', circle.get_circumference())
print('원의 넓이:', circle.get_area())
````

######<ins>데코레이터를 사용한 게터와 세터
케터와 세터 함수를 만드는 일이 많아져서 파이썬에서는 게터와 세터를 쉽게 만들고 사용할 수 있게 하는 기능을 제공합니다. 아래와 같이 변수 이름과 같은 게터 함수를 정의하고 그 위에 `@property`라고 써주시고 변수 이름과 같은 세터 함수를 정의하고 그 위에 `@<게터함수_이름>.setter`이라는 데코레이터를 붙여주세요.

````python
import math

class Circle:
  def __init__(self, radius):
    self.__radius = radius

  def get_circumference(self):
    return 2 * math.pi * self.__radius

  def get_area(self):
    return math.pi * self.__radius**2

  # 게터와 세터 선언
  @property
  def radius(self):
    return self.__radius

  @radius.setter
  def radius(self, value):
    if value < 0:
      raise ValueError("반지름 값은 양수이어야 합니다.")

    self.__radius = value

circle = Circle(10)
print('원의 반지름:', circle.radius)

circle.radius = 2
print('원의 반지름:', circle.radius)

print('인위적으로 예외 발생')
circle.radius = -2
````
