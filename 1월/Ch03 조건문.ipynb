####**시작하기 전에**
Boolean은 불린 또는 불리언이라는 발음으로 부릅니다. 프로그램밍에서는 짧게 Bool이라고 쓰기도 합니다. Boolean Algebra를 부울 대수, Boolean Operator를 부울(불) 연산자로 부릅니다. 부울을 '불'이라고 짧게 발음하기도 합니다.

부울 형 데이터는 True(참)과 False(거짓) 둘 장 하나의 값만을 가질 수 있습니다. 숫자 형 데이터는 무수히 많은 다른 값을 임의로 저장할 수 있었고 문자열 형 데이터도 마찬가지로 무수히 다양한 임의의 문자열을 저장할 수 있었습니다.

### <font color=blue>부울 형 데이터 만들기: 비교 연산자</font>
**비교 연산자**를 이용해서 부울 값을 생성할 수(도) 있습니다. 파이썬에는 6개의 비교 연산자가 있습니다.
<center>

|연산자|설명|연산자|설명|
|:-:|:-:|:-:|:-:|
|==|같다|>|크다|
|!=|다르다| <=| 작거나 같다|
|<|작다|>=|크거나 같다|

</center>

`=`는 할당 연산자입니다.
```python
x = 3
```
`=` 할당 연산자 왼쪽에 있는 것을 오른쪽에 할당(저장)하라는 의미.


````python
print(10 == 100)
result = (10 == 100)
print(result)
````

````python
print(10 != 100)
print( 10 >= 100 )
print( 10 <= 100 )
````

파이썬은 문자열을 가지고 비교 연산을 수행할 수 있습니다.

````python
print('고양이' == '나무')
print('고양이' != '나무')
````

````python
print('고양이' > '나무')
print('고양이' < '나무')
````

위 코드 셀에서 `'고양이' < '나무'`의 연산 결과가 `True`일까요? 문자열에 (크기)비교 연산자를 적용하면 가나다순으로 크기를 따져줍니다.

그런데 이런 연산을 자주 할 일이 있을까 싶습니다.

````python
print('ㅎ' > 'ㄱ')
````

````python
print( 'boy' < 'girl' )
````

###<font color=red>**if 조건문이란?**</font>
`if`문은 조건에 따라 실행할 코드를 나누는데 사용됩니다.
```python

  #조건(condition)이 참이면 여기에 있는 코드를 실행
else:
  #조건(condition)이 거짓이면 여기에 있는 코드를 실행
```

이렇게 조건에 따라 실행될 코드로 이동(점프)하는 것을 조건 분기라고 부릅니다.

### **처음 만나는 ':'**  
<font color=blue>`if 조건_표현'끝에 **반듯이 콜론(:)을 붙여야 합니다**. 그 이유는 나중에 생각해 보기로 하고 지금은 그냥 "파이썬 문법이야"하고 받아들이시면 좋겠습니다. </font>

코드 끝에 콜론(:)을 붙이는 경우가 몇 번 있는데 모두 같은 이유에서 붙입니다. 그 이유가 뭘까요? 생각만 한번 해보고 넘어가겠습니다.

````python
str_n = input('정수를 입력하세요 : ')
int_n = int(str_n)

if int_n > 0:
  print('입력하신 수는 양수입니다.')

if int_n < 0:
  print('입력하신 수는 음수입니다.')

if int_n == 0:
  print('입력하신 수는 0입니다.')
  print('0 맞죠?')

print('판단이 끝났습니다!')
````

###날자/시간 활용하기
시간을 판단 기준으로 삼아 오전이면 어떤 동작을 하고 오후면 또 다른 동작을 하도록 하는 코드를 작성하겠습니다.


````python
import datetime

now = datetime.datetime.now()

print( type(now) )
print(now)
````

<font color = blue>위 코드셀의 1번 줄은 날자와 시간 관련 기능을 수행할 수 있는 `datetime` 모듈을 가지고 오라는 의미입니다. 파이썬을 처음 접하신다면 '모듈'은 일단 함수들의 집합인 라이러리라고 생각하시면 되겠습니다. 컴퓨터 프로그래밍이 처음이라 라이브러리라는 말도 생소하시면 '모듈'은 여러 가지 기능을 가지고 있는 녀석이라고 생각하시면 되겠습니다. 현단에는 `import datatime` 문장을 'date와 time' 관계된 처리를 할 수 있는 연장(기능)이 들어 있는 연장통을 가지고 와라 정도로 생각하시면 되겠습니다.</font>

위 코드셀에서 `now`라는 녀석을 출력해보니 now라는 녀석은 날자와 시간 정보를 가지고 있나 봅니다. 다른 말로 `datetime.datetime.now()`함수가 날자와 시간 정보를 알려주었다는 것을 알수 있습니다.

````python
# 날자
print(now.year,'년')
print(now.month, '월')
print(now.day, '일')

#시간
print(now.hour, '시')
print(now.minute, '분')
print(now.second, '초')
````

위 코드셀 실행 결과로 나타난 시간과 여러 분이 생각하는 시간이 다를 겁니다.

image.png

위 이미지는 파이썬 코드를 제 사무실에서 실행 시킨 결과입니다. 시차가 있네요. 코랩에서 파이썬 코드를 실행하면 이 코드는 코랩 서버에서 실행되고 그 결과를 모니터에 출력하는 것입니다. 따라서 지역(나라)간 시차가 있습니다.

위에서 살펴본 `datetime.datetime.now()` 함수를 이용하여 시간 정보를 구한 다음 12시 이전과 12시 이후 즉 오전일 때와 오후일 때 각기 다른 일을 하도록 코드를 작성해 보겠습니다. "지금이 오전이야? 오후야? 이 조건에 따라 다른 일을 해야해"인 상황인 것이죠.

````python
import datetime

now = datetime.datetime.now()

if now.hour < 12:
  print('오전이네요 아점을 먹고 나서 작업 A를 합니다.')

if now.hour >= 12:
  print('오후네요. 점심을 먹고 나서 작업 B를 합니다.')
````

위 코드 조금 '있어보게' 수정하겠습니다.

````python
import datetime

now = datetime.datetime.now()

if now.hour < 12:
  print('현재 시간은 {}시 {}분입니다. 아점을 먹고 나서 작업 A를 합니다.'.format(now.hour, now.minute))

if now.hour >= 12:
  print('현재 시간은 {}시 {}분입니다. 아점을 먹고 나서 작업 B를 합니다.'.format(now.hour, now.minute))
````

##Ch 03-2 if ~ else와 elif 구문


####**시작하기 전에**
`if`문을 `else`와 짝으로 사용할 수 있는데 문법 형식은 아래와 같습니다.

```python
if 조건식:
  조건식이 참(True)일 때 실행될 코드들
else:
  조건식이 거짓(False)일 때 실행될 코드들
```

"비 온다면 우산을 팔고 그렇지 않다면 신문을 팔아라"를 아래와 같은 형식으로 코드화 할 수 있습니다.

```python
if 비가온다 == True:
  우산을 판다
else:
  신문을 판다.
```

"더우면 보일러를 켜고 그렇지 않으면 보일러를 꺼라"를 코드로 구현하려고 합니다. 덥다 그렇지 안다를 판단하는 조건식을 어떻게 만들어야 할까요? 덥다 그렇지 않다는 사람마다 그리고 그 사람이 입고 있는 옷에 따라 달라질 수 있습니다. 따라서 이렇게 기준을 잡으면 어떨까요. 온도가 26도 이상이면 덥다라고 하고 온도가 26도 미만이면 덥지 않다, 그렇지 않다라고 판단 조건을 잡겠습니다.

```python
if 온도 >= 26:
  보일러 가동 시작
else:
  보일러 가동 중단
```

`if ~ else ~` 형태는 조건에 따라 프로그램 실행이 이것 아니면 저것, 둘 중 한 쪽만 실행하게 됩니다.

```python
if i_num % 2 == 0:
  print('짝수')
else:
  print('홀수')
```

앞에서 임의의 정수를 가지고 짝수인지 홀수인지를 판단하기 위해 `i_num % 2 == 0`이라는 조건식을 사용했었습니다. 짝수가 아니라면 그럼 뭐겠습니까? 홀수겠죠.

`if`(만약) 조건(`i_num % 2 == 0`)이 참(True)이면 `print('짝수')`를 실행하고 그렇지 않다면 `print('홀수')`를 실행하라고 코드를 작성했었네요. '그렇지 않다면'은 조건(`i_num % 2 == 0`)이 참(True)이 아니라면을 뜻하는 것입니다.

위 코드나 아래 코드 두 다 동일한 결과를 출력합니다. 코드 블럭 A는 "`i_num`이 짝수면(_판단_) '짝수'라고 출력하고 그렇지 않으면 '홀수'라고 출력해"를 지시한 것입니다. 코드 블록 B는 "`i_num`이 짝수면(_판단_) '짝수'라고 출력해라. '`i_num` 홀수(_판단_)면 '홀수'라고 출력해라"를 지시한 것입니다.

###if 조건문을 효율적으로 사용하기

아래 코드를 실행시키고 점수를 0점에서부터 100점까지 임의로 입력해 보십시요.

````python
jumsu = input('당신의 점수를 입력하세요(0~100점) : ')
i_jumsu = int( jumsu )

if i_jumsu >= 90:
  print('와우~ 대단하시네요. 어디 가서 자랑하셔도 되겠어요.')
elif i_jumsu >= 80:
  print('이정도면 충분합니다. 다 밥 먹고 살아요.')
elif i_jumsu >= 70:
  print('공부 잘 한다고 다 잘 사는 거 아닙니다.')
elif i_jumsu >= 60:
  print('학교 다닐 때 성적 좋다고 사회에서 성공하나요. 걱정하지 말아요')
elif i_jumsu >= 50:
  print('무슨 일 있어요? 별일 아니죠?')
else:
  print('시대를 앞서가는 혁명가시군요!!')

print('프로그램 종료 - 끝!')
````

위 코드를 실행시킨 후 95를 입력했다면 4번 줄에 보이는 `i_jumsu >= 90` 조건이 참(True)이됩니다. 즉 조건을 만족시킵니다. 그렇다면 `print('와우~ 대단하시네요. 어디 가서 자랑하셔도 되겠어요.')` 코드가 실행됩니다. 그리고 나서는 몇번 줄의 코드가 실행되나요?

`if`만 단독을 코딩할 때가 있었고 `if ~ else ~`와 같이 짝을 이루어 코딩할 때가 있었습니다. `if "코드 블럭 A" else "코드 블록 B"`이와 같은 형태일 때는 두 코드 블록 중 하나의 코드 블록이 실행되었습니다.

```python
if condition1:
  코드 블록 1
elif cindition2:
  코드 블록 2
elif condition3:
  코드 블록 3
#...
else:
  코드 블록 N
```

위와 같이 `if ~ elif ~ elif ~ ... else` 형태일 때도 N개의 코드 블록 중 조건이 만족되는 하나의 코드 블록이 실행되면 전체 `if ~ elif ~ elif ~ ... else`는 끝이 납니다.

위 코드셀에 보인 코드에서 95를 입력했다면 4번 명령 줄의 조건이 만족되는 5번 줄의 명령을 실행하고 `if ~ elif ~ elif ~ ... else`는 끝이 납니다.

####<font color=blue>Quiz</font>
위 Quiz와 동일한 동작을 하되, 1등인 경우 <font color=blue>"받으실 상품은 상품권 5만원, 핸드폰, 노드북입니다"</font>와 같이 출력되도록 코드를 작성하시요. 2등인 경우라면 "받으실 상품은 상품권 5만원, 핸드폰입니다"와 같이 출력되어야 합니다.

````python
s = int( input('몇 등하셨어요? ') )
prize = '받으실 상품은 '

if s == 1:
  prize += '"노트북", '
if s <= 2:
  prize += '"핸드폰", '
if s <= 3:
  prize += '"상품권 5만원"'

if not (s == 1 or s == 2 or s == 3):
  prize += '"꽝"'

print(prize + '입니다.')
````

````python
s = input('몇 등하셨어요? ')
s = int(s)

print('받으실 상품은 ', end='')
prize = ''

if s <= 3:
  prize += '상품권 5만원'
if s <= 2:
  prize +='핸드폰, '
if s == 1:
  prize += '노트북, '

print(prize + '입니다.')
````

`상품명, ` 형태로 인해 쉼표(,)와 스페이스(' ')가 뒤에 보이는데 적절히 제거해야겠습니다.

````python
s = input('몇 등하셨어요? ')
s = int(s)

print('받으실 상품은 ', end='')
prize = ''

if s <= 3:
  prize += '상품권 5만원, '
if s <= 2:
  prize +='핸드폰, '
if s == 1:
  prize += '노트북, '

prize = prize[:-2]
print(prize + '입니다.')
````

###<font color=blue>pass 키워드</font>

친구로부터 전화가 왔습니다. "오늘 저녁 한 잔 할래? 민구도 나오기로 했어." 저는 잠시 생각한 후 "난 **패스**"라고 대답했습니다. 제 말 뜻은 뭔가요?

아래 코드를 실행 후 결과를 보시면서 패스(`pass`)가 어떤 역활을 하는지 생각해 보십시요.

````python
choice = input("""다음 중 지금 제일 먹고 싶은 것을 고르세요.
수박, 참외, 감
무엇을 드시겠습니까? """)

print("당신의 선택은 :'{}'입니다!".format(choice), end='\n\n')

if choice == '수박':
  print('{} 좋죠! 여름에 최고죠.'.format(choice))
elif choice == '참외':
  print('{} 좋죠! 그런데 참외는 우리나라 사람만 먹는다고 하던데요'.format(choice))
elif choice == '감':
  pass
else:
  print('목록에 없는 것을 선택하셨습니다.')

print('끝')
````

위 코드를 다시 실행하고 감을 입력해 보세요.

````python
choice = input("""다음 중 지금 제일 먹고 싶은 것을 고르세요.
수박, 참외, 감
무엇을 드시겠습니까? """)


print("당신의 선택은 :'{}'입니다!".format(choice), end='\n\n')

if choice == '수박':
  print('{} 좋죠! 여름에 최고죠.'.format(choice))
elif choice == '참외':
  print('{}! 좋죠. 그런데 참외를 우리나라 사람만 먹는다고 하던데요'.format(choice))
elif choice == '감':
  pass
  #print('감 좋죠! 너무 많이 먹으면....')
else:
  print('목록에 없는 것을 선택하셨습니다.')

print('끝')
````

키워드 `pass'는 "아무 것도 안함"이라는 뜻으로 쓰이기도 하고 "어떻게 코딩 해야할지 아직 생각 중이다. 나중에 여기에 코드를 넣을테니 일단은 "<font color=blue>**아무 것도 하지 말고**</font> 넘어가자"라는 의도입니다. 전자의 의미는 "코드 완성 한거야. 이 때 아무 것도 안할거야."이며 후 자는 "일단 그냥 넘어가자"라는 뜻입니다.

지난 주말에 식물원에 다녀왔습니다. 입장료를 받더군요.
만 7세 미만은 무료이고 만 7세부터 만 12세까지는 2,000원
만 12세 이상부터 만 18세 미만까지는 4,000원,
만 18세부터 만 64세 미만까지는 8,000원,
만 65세 이상은 무료였습니다.

<center>

|나이|요금|
|:-:|:-:|
|0~7세 미만|무료|
|7~12세 미만|2,000|
|12~18세 미만|4,000|

|65세 이상| 무료|

만 나이를 입력하면 나이에 맞는 이용료를 출력하는 프로그램을 작성하세요.


````python
age = int( input('당신의 만 나이를 입력하세요 : ') )

if age < 7:
  price = '무료'
elif age <= 12:
  price = 2000
elif age < 18:
  price = 4000
elif age < 64:
  price = 8000
else:
  price = '무료'

print('만 나이가 {}이신 당신의 요금은 {}입니다.'.format(age, price))
````

90\~100은 A, 즉 100점을 포함한 90점대

80\~90 미만까지는 B, 즉 80점대

60\~70 미만까지는 D, 즉 60점대

0\~60 미만은 F

|점수|학점|
|:-:|:-:|
|90이상 ~ 100이하|A|
|80이상 ~  90미만|B|
|70이상 ~  80미만|C|
|60이상 ~  70미만|D|
|0이상  ~  60미만|F|



실수형태로 점수를 입력 받아서 A, B, C, D, F를 출력하는 코드를 작성하시오.  
예를 들어, 점수가 82.4인 경우 B 학점입니다.


````python
int(92.4/10)
````

````python
jumsu = float( input('점수를 입력하세요(실수 : 0~100) : ') )

idx = int( jumsu/10 )

if idx < 6:
  grade = 'F'
elif idx < 7:
  grade = 'D'
elif idx < 8:
  grade = 'C'
elif idx < 9:
  grade = 'B'
else:
  grade = 'A'

print(grade)
````
