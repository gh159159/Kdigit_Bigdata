## 문제 0
아래 코드의 동작을 분석하시오.
```python
class Stack():
  def __init__(self):
    self.lst = []
  def push(self, element):
    self.lst.append(element)
  def pop(self):
    return self.lst.pop()

stack = Stack()
stack.push(10)
stack.push(20)
stack.push(30)
print( stack.pop() )
print( stack.pop() )
```



아래 코드가 실행되면 예외(exception)이 발생할 것입니다.
```python
class Stack():
  def __init__(self):
    self.lst = []
  def push(self, element):
    self.lst.append(element)
  def pop(self):
    return self.lst.pop()

stack = Stack()

for idx in range(5):
  stack.push(idx)

for _ in range(10):
  print( stack.pop() )
```

`Machine` 클래스가 어떻게 정의 된었기에 아래와 같은 결과 나왔을까요? `Machine` 클래스를 정의한 코드를 추측해서 작성해 보세요.

```python
class NewMachine(Machine):
  def __init__(self):
    super().__init__()
    self.setting()
  
  def run(self):
    print('위이잉~~')

  def setting(self):
    print('징~ 징~ 사전 설정 중입니다.')
    print('사전 설정 완료되었습니다.')
  
machine = Machine()
machine.run()

print('-'*32)

new_machine = NewMachine()
new_machine.run()
```

##문제 5
아래 코드를 참고 하십시요.
```python
class Tiguan():
  def __init__(self, name):
    self.name = name

  def run(self):
    print('꼼지란~ 꼼지락~')

  def hello(self):
    print('안녕하세요~ 제 이름은 {}이구요 하늘을 날 수 있답니다.'.format(self.name))    # look at me

class Wing():
  def fly(self):
    print('푸드덕~ 푸드덕~')

class FlyingTiguan(Tiguan):
  def __init__(self, name, color):
    super().__init__(name)
    self.color = color
    self.hello()
    self.wing = Wing()       # look at me

  def shout(self):
    print('My parent is a Tiguan!')

  def run(self):
    print('하나 둘~ 하나 둘~')
```

FlyingTiguan 객체가 날개 짓을 천천히 또는 빠르게 움직이는 동작을 구현하려고 합니다. 날개짓을 빠르게 또는 천천히 한다는 동작을 아래 코드와 같이 구현하려고 합니다.

```python
import time
# 날개짓을 빠르게
for i in range(5):
  print('푸드덕~ 푸드덕~')
  time.sleep(0.1)

print('-'*32)

# 날개짓을 천천히
for i in range(5):
  print('푸드덕~ 푸드덕~')
  time.sleep(1.5)
```

맨 위에 제시한 코드를 수정해서 FlyingTiguan 객체가 날개짓을 빠르게 하거나 천천히 할 수 있도록 만드십시요.
