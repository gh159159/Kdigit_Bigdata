###<font color=blue>가변 매개변수</font>

앞서 살펴본 함수는 함수 선언할 때의 매개변수 개수와 함수를 호출할 때의 매개변수의 개수가 같아야 합니다. 적어도 안되고 많아도 안됩니다. 그런데 그동안 많이 사용했던 `print()` 함수는 매개변수를 1개를 사용해도 되고 2개를 사용해도 되고... 아무런 문제가 없었습니다.

````python
print('Hello')
print('Hello', 'world', '!')
````

 이와 같이 **매개변수를 주는 데로 받을 수 있는 함수를** <font color=blue>**가변 매개변수함수**</font>라고 합니다. 매개변수를 주는 데로 받을 수 있다는 말은 매개 변수의 개수가 변할 수 있다는 의미입니다. 가변 매개변수 함수는 다음과 같은 형태로 만듭니다.

```python
def function_name(일반_매개변수1, 일반_매개변수2, ..., *가변_매개변수):
  #statement(문장)
```
<font color=blue>**가변 매개변수**</font>라는 것을 명시하기 위해 매개변수 앞에 `*`를 붙여 표시합니다. <font color=blue>가변 매개변수를 사용할 때는 아래 두 가지 제약</font>이 있습니다.
1. 가변 매개변수 뒤에는 일반 매개변수를 위치시킬 수 없음.
2. 가변 매개변수는 하나만 사용할 수 있음.


위와 같은 제약 사항이 없으면 가변 매개변수가 어느 것인지 알수 없기 때문에 만들어진 제약입니다. 예를 들어 아래 코드와 같은 함수 선언이 있다고 가정하겠습니다.

```python
def what_eat(n, *data):
  # statement(문장)
```

위 함수 선언에서 매개변수 `n`은 <font color=blue>**일반 매개변수**</font>입니다. `data`가 가변 매개변수입니다. 가변 매개변수 `data`뒤에 일반 매개변수가 없습니다. 그리고 가변 매개 변수는 하나만 있습니다.

````python
def print_n_times(n, *values):
  for i in range(n):
    for value in values:
      print( '{} '.format(value), end='' )
    print()

print_n_times(3, '안녕하세요', '즐거운', '파이썬 프로그래밍')
````

````python
def print_n_times(n, *values):
  for i in range(n):
    for value in values:
      print( '{} '.format(value), end='' )
    print()

  print("'values'의 자료형은 :", type(values))

print_n_times(3, '안녕하세요', '즐거운', '파이썬 프로그래밍')
````

````python
def print_n_times(n, *datas):
  print(datas)                        # 여기가 변경된 부분
  for i in range(n):
    for data in datas:
      print( '{} '.format(data), end='' )
    print()

print_n_times(2, '사과', '수박', '참외')
print('--- 구분선 ---')
print_n_times(3, '수박', '참외', '딸기', '귤')
````

만약 위에서 말한 "가변 매개변수 뒤에 일반 매개 변수를 위치시킬 수 없음" 제약이 없다면 함수 선언 형태를 아래와 같이 했을 수 있습니다.

```python
def function_name(*가변_매개변수, 일반_매개변수1, 일반_매겨변수2, ...):
  #statement(문장)
```

그리고 `print_n_times('사과', '수박', '참외', 3)`과 같이 호출했을 것입니다. 이 경우 이 4개의 매개변수 중 몇 개가 가변 매개변수에 해당하고 몇 개가 일반 매개변수인지 구분하기 어렵습니다. 가변 매개변수에 해당하는 것이 `'사과'`, `'수박'`까지인지 `'사과'`, `'수박'`, `'참외'`까지인지 모호합니다.




아래 코드는 가변 매개변수와 일반 매개변수가 같이 사용되고 있는 예입니다. 일반 매개변수가 가변 매개변수 보다 앞에 위치해야 한다는 제약을 위반하고 있습니다.

````python
def print_n_times(*datas, n):
  print(datas)                        # 여기가 변경된 부분
  for i in range(n):
    for data in datas:
      print( '{} '.format(data), end='' )
    print()

print_n_times('사과', '수박', '참외', 2)
````

위 코드에서는 함수에 4개의 값을 매개 변수로 전달하고 있습니다. 총 4개 중 몇 개를 가변 매개 변수에 할당해야할지 파이썬 입장에서는 모호합니다.

###기본 매개변수(Default parameters)

그동안 편하게 써온 `print()` 함수는 다음과 같은 형태로 매개변수를 받는다고 합니다.
```python
print(value, ..., sep=' ', end='\n', file=sys.stdout, flush=False)
```

위 `print()` 함수의 `value`가 **가변 매개변수**입니다. 앞에서 가변 매개변수는 일반 매개변수 앞에 올 수 없다고 말씀드렸습니다. 그렇다면 `sep`, `end`, .., `flush`는 일반 매개변수가 아니다? `sep`, `end`, .., `flush`는 <font color=blue>**기본 매개변수**</font>입니다. 기본 매개변수는 `매개변수 = 값` 형태이며, 해당 함수를 호출할 때 기본 매개변수 값을 지정하지 않을 경우 `매개변수=값`에 의해 값이 자동으로 할당됩니다.


`print()`함수에서 `end`라는 매개 변수가 보입니다. 이 매개변수는 `end='\n'`과 같이 `end`의 값은 `\n`으로 내정되어 있습니다. 매개변수 end에 우리가 별도로 값을 할당 하지 않는다면 내정된 값, `\n`이 할당됩니다.  


````python
print('hello world!')
print('my nickname is ship')
````

`print()`함수의 기본 매개변수 `end`에는 print 동작 후 마지막(끝, end)에 어떤 문자를 출력할 것인지 지정하는 변수입니다. 그런데 `print()`함수를 호출하면서 `end` 기본 매개변수에 아무런 값을 전달하지 않으면 사전에 설정된 기본값인 `'\n'`을 출력합니다. 그래서 `print()`함수는 주어진 문자열을 출력한 후 다음 출력 위치를 한 줄 아래로 이동했던 것입니다.



````python
print('hello', 'world!',sep='@', end=' ')
print('my nickname', 'is ship')
````

함수에서 <font color=blue>기본 매개변수를 사용할 때 제약 조건이 하나 있습니다. </font>
- 기본 매개변수 뒤에는 일반 매개변수가 올 수 없습니다.

오늘 수업 끝나면 저녁에 특별한 계획이 있나요? 전 가능하다면, 지인들과 가볍게 한 잔 하고 싶습니다. 마음만.... ㅎㅎ

````python
def first_order(anju, sul = '참이슬 프레쉬'):
  # 매개변수 2개를 받음
  print(anju, end=' ')
  print('그리고', end = ' ')
  print(sul, end=' ')
  print('주세요')

first_order('삼겹살')
first_order('삼겹살', '맥주')
````

위에 보인 `first_order()` 함수는 일반 매개변수와 기본 매개변수가 같이 사용된 경우입니다. 이 경우 매개변수의 순서는 일반 매개 변수가 먼저 오고 기본 매개 변수를 둡니다.

- 일반 매개변수1, 일반 매개변수2, ... 가변 매개변수
- 일반 매개변수1, 일발 매개변수2, ... 기본 매개변수1, 기본 매개변수2, ...

````python
def first_order(anju, sul = ['참이슬 프레쉬', '테라']):
  # 매개변수 2개를 받음
  print(anju, end=' ')
  print('그리고', end = ' ')
  for s in sul:
    print(s, end=' ')
  print('주세요')

first_order('삼겹살')
first_order('삼겹살', '맥주')
first_order('참치회', ['소주', '맥주'])
````

def function_name(매개변수):
  변수 = 초깃값

````python
# 함수를 선언합니다.
def sum_all(start, end):
  # 변수를 선언합니다.
  output = 0

  # 여러 가지 처리 - 반복문으로 숫자를 더함
  for i in range(start, end+1):
    output += i

  # 변수에 저장된 값을 반환
  return output

print('1 to 100:', sum_all(1 ,10))
print('50 to 100:', sum_all(50, 100))
````

````python

# 함수를 선언합니다.
def sum_all(start=0, end=100, step = 1):
  # 변수를 선언합니다.
  output = 0

  # 여러 가지 처리 - 반복문으로 숫자를 더함
  for i in range(start, end+1, step):
    output += i

  # 변수에 저장된 값을 반환
  return output

print('1 to 100:', sum_all(1 ,10, 1))   # start = 1, end = 10, step = 1
print('50 to 100:', sum_all(50))        # start = 50, end = 100, step = 1
print( sum_all(start=50, end=100, step=1) )
print( sum_all(start=50, step=2, end=100) )
````

#####Quiz

식 (5-1)은 식 (5-2)와 동일한 값을 갖는다는 점을 고려해서 저도 코드를 작성해 보겠습니다.

$$n! = 1 \times ... \times (n-2) \times (n-1) \times n \tag{5-2}$$


````python
def factorial(n):
  s = 1
  for nn in range(1, n+1):                # n! = 1 x 2 x ... x (n-2) x (n-1) x n
    s *= nn
  print(s)


factorial(4)  # 1 x 2 x 3 x 4 = 24
````

````python
def factorial(n):
  s = 1
  for nn in reversed( range(1, n+1) ):    # n! = n x (n-1) x (n-2) x ... x 1
    s *= nn
  print(s)


factorial(4)  # 4 x 3 x 2 x 1 = 24
````

####<ins>재귀 함수로 팩토리얼 구하기</ins>
재귀(recursion)란 '자기 자신을 호출하는 것'을 의미합니다. 식(5-1)을 식(5-3)과 동일한 의미죠.

$$n! = n \times (n-1)! \tag{5-3}$$

그렇다면 앞서 우리는 만든 `factorial()`함수를 이용하여 $n!$ 계산을 위한 코드를 아래와 같이 만들 수 있겠네요.

```python
n * factorial(n-1)
```
그럼 다시 $(n-1)!$ 계산을 위한 코드를 아래와 같이 만들 수 있겠습니다.

```python
(n-1) * factorinal(n-2)
```

$$f(4) = 4 \times f(3)\\  
       = 4 \times 3 \times f(2)\\  
       = 4 \times 3 \times 2 \times f(1)\\  
       = 4 \times 3 \times 2 \times 1 \times 1$$

이쯤에서 아래 코드 셀의 코드를 살펴보십시요.



````python
def factorial(n):
  if n == 1:      # n이 0이라면 1을 리턴
    print('{} = '.format(n), end='')
    return 1
  else:
    print('{}x'.format(n), end='')
    return n * factorial( n - 1)

print(factorial(1))
print(factorial(2))
print(factorial(3))
print(factorial(4))
````

**리스트 평탄화**는 중첩된 리스타가 있을 때 중첩을 모두 제거 하고 풀어서 1차원 리스트로 바꾸는 것을 말합니다. 다음과 같이 세번 중첩된 리스트를 평탄화하려면 어떻게 해야할까요?

<center>

`[[1, 2, 3,], [4, [5, 6]], 7, [8, 9]]` →`[1, 2, 3, 4, 5, 6, 7, 8, 9]`

</center>

두번 중첩된 리스트를 평탄화하는 것은 비교적 쉬운 문제이지만 몇 번 중첩되었는지 모르는 리스트를 평탄화하는 작업은 조금 어렵습니다. 먼저 두 번 중첩된 리스트 하나를 입력받아 평탄화하는 함수를 만들어 보겠습니다.


````python
def flatten(data):
  output =[]
  for item in data:
    if type(item) == list:
      output += item
    else:
      output.append(item)
  return output

lst = [[1, 2, 3],4, [5, 6], 7]


print('원본 :', lst)
print('변환 :', flatten(lst))
````

이번에는 몇 번 중첩되었는지 모르는 리스트를 평탄화 하는 코드를 살펴보겠습니다.

````python
def flatten(data):
  output =[]
  for item in data:
    if type(item) == list:
      output += flatten(item)
    else:
      output.append(item)
  return output


lst = [[1, 2, 3,], [4, [5, 6]], 7, [8, 9]]
print('원본 :', lst)
print('변환 :', flatten(lst))
````

####<ins>람다의 개념</ins>
매개변수로 함수를 전달하기 위해 해당 **함수를 코딩하는 하는 것도 번거롭고 코딩 공간 낭비**라는 생각을 하는 사람들이 많아졌다고 합니다. 그래서 심플한 방법을 생각하게 되었고 그래서 등장한 개념이 람다(lambda)입니다. 람다라는 개념은 '간단한 함수는 심플한 문법으로 쉽게 만들 수 있게 하자'입니다.

```python
lambda 매개변수: 리턴값
```

````python
def power(x):
  return x*x

x = range(10)
y = map(power, x)

print(list(y))
````

````python
power = lambda x : x*x

x = range(10)
y = map(power, x)

print(list(y))
````

lambda? 감동적이지도 않고 재미있지도 않으시죠?

````python
x = range(10)
y = map(lambda x:x*x, x)

print(list(y))
````

여전히 재미있지는 않지만 살짝 감동적이지 않습니까?  매개변수로 함수를 전달하기 위해 해당 함수를 코딩하는 하는 것도 번거롭고 코딩 공간 낭비라는 생각을 하는 사람들에게 도움이 되는 기능일지 모르지만 우리에게 아직까지는 큰 감동은 아니네요. ㅎㅎ

함수 선언할 때 사용하는 키워드 `def`와 값을 반환할 때 쓰는 `return`을 코딩하지 않아도 되네요.

**\[제너레이터(generator)란?\]**

**제너레이터(generator)**는 파이썬의 특수한 문법입니다. 제너레이터는 이터레이터(iterator)를 직접 만들 때 사용하는 코드입니다. 다른 측면에서 제너레이터를 설명하자면, 일반적인 함수는 값을 한번 반환합니다. 그 값은 정수일 수도 있고 리스트이거나, 딕셔너리거나 튜를일 수도 있습니다. 함수가 데이터를 한번만 반환하는 것이 아니라 지속적으로 계속 반환할 수 있었으면 좋겠다는 발상에서 만들어진 것이 제너레이터입니다.


제너레이터는 이터레이터와 마찬가지로 next() 함수 호출 시 그 값을 차례대로 얻을 수 있습니다. 제너레이터에서는 차례대로 결과를 반환하고자 return 대신 yield 키워드를 사용합니다.

````python
def mygen():
  yield 'a'
  yield 'b'
  yield 'c'

g = mygen()
print( type(g) )

print( next(g) )
print( next(g) )
print( next(g) )
````

#####**\[제너레이터 표현식(expression)\]**
이래 코들 먼저 보십시요.


````python
def mygen():
  for i in range(1, 7):
    result = i*i
    yield result

gen = mygen()

print(next(gen))
print(next(gen))
print(next(gen))

for item in gen:
  print(item)
````

제너레이터는 def를 이용한 함수로 만들수도 있지만, 다음과 같이 튜플 표현식으로 좀 더 간단하게 만들 수도 있습니다.

````python
gen = (i*i for i in range(1,7))

print( type(gen) )

print( next(gen) )
print( next(gen) )
print( next(gen) )
````

위 표현식(expression)은 `mygen()` 함수로 만든 제너레이터와 완전히 똑 같습니다.여기서 사용한 표현식은 리스트 내포(list comprehension) 구문과 비슷합니다만 리스트가 아닌 튜플을 이용한다는 점이 다릅니다. 이와 같은 표현식을 제너레이터 표현식(generator expression)이라고 합니다.

````python
gen = [i*i for i in range(1,7)]      # 이것은 리스트 내포(list comprehension)입니다.

print( type(gen) )
````

#####**\[제너레이터의 쓰임새\]**
제너레이터를 어떤 상황에서 쓰면 좋을까요?

````python
import time

def longtime_job():
  print('job start')
  time.sleep(1)
  return 'done'

list_job = iter([longtime_job() for i in range(5)])

print(type(list_job))

print( next(list_job) )
print( next(list_job) )
print( next(list_job) )
print( next(list_job) )
print( next(list_job) )
````

위 코드 셀의 8번 줄에서 리트스 내포로 리스트를 만드는데 `for in range(5)`에 의해 요소가 5개인 리스트를 만들게 됩니다. 리스트의 각 요소는 `longtime_job()` 함수의 반환값이 됩니다. 즉 리스트 내포로 만들어진 리스트는 `['done', 'done', 'done', 'done', 'done']`가 됩니다. 이 리스트를 `iter()`함수를 이용해서 이터레이터로 변환합니다. 따라서 `list_job`은 (리스트) 이터레이터가 됩니다.

````python
import time

def longtime_job():
  print('job start')
  time.sleep(1)
  return 'done'

list_job = (longtime_job() for i in range(5))

print(type(list_job))

print( next(list_job) )
````

`iter([longtime_job() for i in range(5)])`를 제너레이터 표현식 `(longtime_job() for in range(5))`로 변경했습니다. 실행 방식이 위 코드와 전혀 다릅니다. 리스트 내포에 의해 리스트의 각 요소의 값은 `longtime_job()` 함수를 호출하여 반환된 값을 각 요소에 할당하여 리스트를 완성합니다. 그러나 제너레이터 표현식 `(longtime_job() for in range(5))`은 각 요소 값을 결정하는 시기가 앞선 경우와 다릅니다. `next(list_job)`가 실행될 때 각 요소의 값을 그 때 결정합니다. 즉 `next(list_job)`가 실행될 때 `longtime_job()`함수가 실행됩니다.

````python
import time
import datetime

def longtime_job():
  print('{} job start'.format(datetime.datetime.now()))
  time.sleep(1)
  return 'done'

list_job = (longtime_job() for i in range(5))

print(type(list_job))

print( next(list_job) )
````

######Quiz
그런데 다음과 같이 딕셔너리를 활용해서 책에 대한 정보를 표현했을 때 '가격이 가장 싼 책'과 가격이 가장 비싼 책'을 찾고 싶아면 어떻게 해야할까요? 코드를 작성해 보시죠.

````python
books = [
          {
              '제목':'혼자 공부하는 파이썬',
              '가격':18000

          },
          {
              '제목':'혼자 공부하는 머신러닝 + 딥러닝',
              '가격':26000
          },
          {
              '제목':'혼자 공부하는 자바스크립트',
              '가격':24000
          }
        ]
````

````python
min_price = 1000000
max_price = 0
min_idx = None
max_idx = None

for idx, item in enumerate( books ):

  if item['가격'] < min_price:
    min_price = item['가격']
    min_idx = idx

  if item['가격'] > max_price:
    max_price = item['가격']
    max_idx = idx

print('가장 싼 책은 "{}"이고 가격은 {}입니다.'.format(books[min_idx]['제목'], min_price))
print('가장 비싼 책은 "{}"이고 가격은 {}입니다.'.format(books[max_idx]['제목'], max_price))
````

`min()`함수와 `max()`함수는 어떤 값을 기준으로 최초, 최대를 찾을지 설정할 수 있습니다. `key`라는 매개변수에 지정을 하면 됩니다. 이를 활용하여 딕셔너리에 있는 `가격`을 `min()` 함수 또는 `max()` 함수의 매개변수 `key`에 할당해서 판단 기준을 가격으로 설정할 수 있습니다.


````python
def get_price(book):
  return book['가격']

print( min(books, key = get_price) )
print( max(books, key = get_price) )
````


그래서 요점은 위와 같은 방식을 적용했을 때, 보다 편리하게 **"딕셔너리에 있는 특정 키(key)의 최솟값과 최댓값을 기준으로 딕셔너리 요소를 추출할 수 있다"**입니다.

람다(lambda)를 이용하여 위 코드를 수정해 보십시요.

````python
def get_price(book):
  return book['가격']

print( min(books, key = lambda book : book['가격']) )
print( max(books, key = lambda dic : dic['가격']) )
````

````python
lst = [ 1, 2, 9, -1, 7]
lst.sort()
print(lst)
````

````python
lst = [ 1, 2, 9, -1, 7]
lst.sort(reverse=True)
print(lst)
````

````python
books = [
          {
              '제목':'혼자 공부하는 파이썬',
              '가격':18000

          },
          {
              '제목':'혼자 공부하는 머신러닝 + 딥러닝',
              '가격':26000
          },
          {
              '제목':'혼자 공부하는 자바스크립트',
              '가격':24000
          }
        ]

# 기본 설정인 오름차순 정렬
books.sort( key = lambda book:book['가격'] )

for book in books:
  print(book)
````

#####좀 더 알아보기(5) - 함수의 값 복사와 레퍼런스 복사
파이썬은 함수를 호출할 때 함수 내부 코드를 실행하기 위해 함수 내부의 변수를 저장할 스택을 추가로 만듭니다. 아래 코드를 실행하면 메모리에 2개의 스택이 만들어지는데 하나는 **함수 스택** 다른 하나는 **전역 스택(global stack)**이라고 합니다.

```python
def ftn(b):
  c = b + 10

a = 10
ftn(a)
```


위 코드에서 `ftn(a)`로 `ftn()`함수를 호출합니다. <font color=blue>`ftn()`함수의 매개변수로 `a`를 전달했기 때문에 **`a`에 들어 있는 값**이 `ftn()`함수로 전달됩니다. 매개변수 `a`에 저장된 값 10은 `b`에 **복사**되어 `ftn()`함수에서 `b`라는 이름으로 활용됩니다.</font> 여기서 핵심 사항은 `a`와 `b`는 전혀 다른 것입니다. 숫자 10을 저장하기 위해 메모리 공간이 필요한데 `a=10`의 의미는 숫자 10을 메모리에 저장하고 그 위치를 `a`라고 부르겠다는 의미가 포함되어 있습니다. `a`이라고 명명된 메모리 위치와 `b`라고 명명된 메모리 위치는 다른 곳입니다.

````python
def change(a):
  a = 20

a = 10
print(a)
change(a)
print(a)
````

위 코드 셀의 5번, 7번 줄에 있는 `print()`함수로 인해 `a`에 저장된 값이 2번 출력되었습니다. 5번 줄에 있는 `print(a)`함수로 인해 `a`에 저장된 값 10이 출력되었고 6번 줄에서 `change(a)`함수를 호출했습니다.`change()`함수에서는 `a`에 20을 저장했습니다. 7번 줄에 있는  `print(a)`함수로 인해 'a'에 저장된 값을 출력했습니다. 그런데 그 값이 20이 아니라 10입니다.

4번 줄에서 선언 `a`를 선언하면서 10을 저장하게 만들었습니다. 그리고 6번 줄에서 `change(a)`함수를 호출하여 'a'가 가지고 있는 값을 복사 해서 이름은 같지만 데이터를 보관하는 메로리 주소가 다른 'a'에 그 값을 넣은 것입니다. `change()` 함수 내부의 'a'와 `change()` 함수 밖의 `a`는 비록 이름은 같지만 서로 다른, 데이터를 보관하는 메모리 위치가 다릅니다. 이름만 같은뿐 서로 다른 변수입니다.



#####좀 더 알아보기(5) - 함수의 값 복사와 레퍼런스 복사 - <ins>객체 자료형 복사</ins>
다음 코드의 실행 결과를 예측해 본 후 코드를 실행시켜 얻은 결과와 비교해 보세요.

```python
def change(a):

```

````python
def change(a):
  a.append(4)

a = [1, 2, 3]
print(a)
change(a)
print(a)
````

이전 경우와 달리 매개 변수로 전달한 `a`의 값이 변경되었습니다. 위 코드 셀에서 ` a`는 힙에 저장된 리스트 위치를 나타내는 주소값을 가지고 있습니다. 리스트 형 데이터에 대한 주소값이 저장된 `a`이 `change()`에 매개변수로 전달한 것입니다. `change()`함수에서는 매개변수로 전달되어 온 값(주소)을 (이름은 같지만 다른 변수)'a'로 복사한 것입니다. `change()` 함수 안에 있는 'a'를 가지고 힙 영역에 있는 동일한 리스트에 접근할 수 있습니다. 따라서 'change()` 함수 안에 있는 'a.append(4)'를 실행시켜 힙에 있는 리스트를 변경한 것입니다.
