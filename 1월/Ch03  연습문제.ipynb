##문제 8

사용자로부터 임의 수를 입력 받아서 소수점 이하 숫자가 없는 정수이면 해당 수를 출력하고 실수이면 소수점 이하 숫자만 출력하는 코드를 작성하시오.


### 풀이 8-1
임의의 수를 입력 받는 다는 것은 (정수 포함하여) 실수를 입력 받는다는 뜻.
정수면 해당 수를 (그대로) 출력하고 실수라면 소수점 이하 숫자만 출력하라는 것은 먼저 임의로 입력 받은 숫자가 소수점 이하 숫자가 없는 정수인지 아니면 정수를 제외한 (소수점 이하 숫자가 있는) 실수인지를 파악해야 함.



````python
x = 23.2
print(x % 1)
# 임의의 수 x가 소수점 이하 숫자를 갖는다면 1로 나누었을 때 나머지가 발생.
# 임의의 수 x가 소수점 이하 숫자가 없는 정수라면 1로 나누었을 때 너머지는 0.

x = 23
print(x % 1)
````

실수에서 소수점 이하 숫자만 구하려면 임의의 실수(`x`)에서 임의의 실수(`x`)를 1로 나눈 몫을 빼면 됨.

````python
x = float( input('임의의 숫자 입력 : ') )
d = x // 1         # 임의의 실수 x를 1로 나눈 몫
s = x % 1          # 임의의 실수 x를 1로 나눈 후 나머지.
                   # 나머지가 0이 아닌 경우 임의의 실수 x는 1로 나누어 떨어지지 않다는 것을 알 수 있음
                   # 이는 곧 소수점 이하 값이 있다고 판단!

if s:
  print('{:.4f}'.format(s))
else:
  print(int(d))
````

### 풀이 8-3

````python
x = float( input('임의의 숫자 입력 : ') )

str_x = str(x)               # 임의의 실수를 문자열로 변경
idx = str_x.find('.')        # 문자열에서 '.'의 인덱스(위치) 파악
xx = str_x[:idx]             # 문자열에서 '.' 이전 문자열만 슬라이싱 --> 소수점 이하 버림.
xx = int(xx)                 # 소수점 이하 숫자가 없는 형태의 문자열을 정수 또는 실수로 변환

if x == xx:                  # 사용자가 입력한 숫자와 사용자가 입력한 숫자에서 소수점 이하를 버린 숫자가 같다면(3.0 == 3) 정수
  print(xx)
else:                        # 실수라면
  print('{:.4}'.format( x - (x//1)) )
````
